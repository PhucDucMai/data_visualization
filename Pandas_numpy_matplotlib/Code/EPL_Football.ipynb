# Map dữ liệu của cột FTR sang dữ liệu số: 1. Hòa 2. Khách thắng 3.Nhà thắng

```python
mask = {'D':1 , 'A':2 , 'H':3}
df_epl_copy['FTR'] = df_epl_copy['FTR'].map(mask)
```

```python
x = df_epl.groupby(['Season' , 'FTR'])['FTR'].count()
```

```python
y = pd.DataFrame(x)
```

```python
y.head()
```

```python
y.rename(columns = {'FTR':'Way'} , inplace = True)
```

```python
y.head()
```

```python
y = y.reset_index()
```

```python
y.head()
```

```python
lst_home = []
lst_away = []
lst_season = []
lst_percent = []
d = 0
a = 1
h = 2
while(a < len(y['FTR']) and h < len(y['FTR'])):
    points_away = y['Way'][a] * 3
    points_home = y['Way'][h] * 3
    lst_home.append(points_home)
    lst_away.append(points_away)
    lst_season.append(y['Season'][a])
    percent_win = (y['Way'][h] / (y['Way'][d] + y['Way'][a] + y['Way'][h])) * 100
    lst_percent.append(percent_win)
    a = a + 3
    h = h + 3
    d = d + 3
print("\tTổng số điểm của từng mùa kiếm được từ sân nhà là:")
for i in lst_home:
    print(i)
print("\tTổng số điểm của từng mùa kiếm được từ sân khách là:")
for j in lst_away:
    print(j)
```

# Vẽ biểu đồ

```python
arr_home = np.asarray(lst_home)
arr_away = np.asarray(lst_away)
arr_percent_home_win = np.asarray(lst_percent)
```

```python
arr_home.shape
```

```python
arr_season = np.asarray(lst_season)
```

```python
arr_season.shape
```

```python
fig = plt.figure(figsize=(8,10))
plt.plot(arr_season, arr_home , label='Home')
plt.plot(arr_season , arr_away , label='Away')
plt.xticks(rotation=90)
plt.legend()
plt.show();
```
